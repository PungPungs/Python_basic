```python
df1 = pd.DataFrame({
    'col1' : ['사과','배','감','수박','메론'],
    'col2' : [500,1000,2500,5000,3000],
},index=[0,1,2,3,4])
df2 = pd.DataFrame({
    'col1' : ['수박','멜론','딸기','키위','오랜지'],
    'col2' : [5000,3000,1000,600,700],
}, index=[3,4,5,6,7])
```

```python
pd.concat([df1,df2],ignore_index=False)
```

```python
pd.concat([df1,df2],ignore_index=True)
```

```python
pd.concat([df1,df2],axis=1)
```

```python
df3 = pd.DataFrame({
    'item' : ['item0','item1','item2','item3'],
    'count' : ['count0','count1','count2','count3'],
    'price' : ['price0','price1','price2','price3']
},index=[0,1,2,3])

df4 = pd.DataFrame({
    'item' : ['item2','item3','item4','item5'],
    'count' : ['count2','count3','count4','count5'],
    'price' : ['price2','price3','price4','price5'],
    'var' : ['var2','var3','var4','var5']
},index=[2,3,4,5])
```

```python
pd.concat([df3,df4],join='outer')
# outer 는 합집합, inner 은 교집합, 컬럼에 없는 것도 합침
```

```python
pd.concat([df3,df4],join='inner')
# 컬럼에 없는 건 안합침
```

```python
# 중복이 있으면 오류 발생, verify_integrity = True 면
pd.concat([df3,df4],verify_integrity=True)
```

```python
# merge 함수는 on 으로 어떤 컬럼을 통해 merge 할 것인지, how는 어떤 파라미터로 merge 할 것인지
pd.merge(df3,df4,on='item')
```

```python
# inner 일치하는 값이 있는 경우에만 데이터 가져옴
# left 왼쪽열 기준으로 오른쪽 데이터프레임 병합
# right 오른쪽열 기준으로 왼쪽 데이터프레임 병합
# outer left 와 right를 한 후 두 개를 합침
pd.merge(df3,df4,on='item',how='inner')
```

```python
pd.merge(df3,df4,on='item',how='left')
```

```python
pd.merge(df3,df4,on='item',how='right')
```

```python
pd.merge(df3,df4,on='item',how='outer')
```
